```python
import pandas as pd
df = pd.read_csv(r'C:\Users\P siva kumar\Downloads\icc_wc_23_bat.csv')
df.sample(10)
df.shape
```

Highest runs scored in one single match

```python
highest_runs_one_match= df.loc[df['runs'].idxmax()]
highest_runs_one_match['player']
```

Total runs scored by all the players from highest to lowest

```python
players_sorted = df.groupby('player')['runs'].sum().sort_values(ascending = False).reset_index()
players_sorted
top_scorers = players_sorted.head(10)
top_scorers
```

Top Scorers - Plot using Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12, 6))
ax1 = top_scorer_plot = sns.barplot(x='player', y='runs', data = top_scorers)
plt.xticks(rotation=45)
plt.xlabel('Player')
plt.ylabel('Runs Scored')
plt.title('Runs Scored by Players')
plt.xticks(rotation=20)
plt.show()
```

Top Scorers Plot using Matplotlib

```python
import matplotlib.pyplot as plt
ax1 = plt.bar(top_scorers['player'],top_scorers['runs'],width = 0.4)
plt.subplots_adjust(left=0.1, right=0.9, bottom=0.1, top=0.9, wspace=0.4, hspace=0.4)
plt.xlabel('Players')
plt.ylabel('Runs')
plt.title('Top Scorers')
plt.xticks(rotation=30)
plt.show()
```

Number of times a player was not out

```python
def count_not_out(df):
    players_count = df[df['dismissal']=='not out'].groupby('player').size().to_dict()
    return players_count
```

```python
player_not_out_counts = count_not_out(df)
print(type(player_not_out_counts))
```

```python
for player, count in player_not_out_counts.items():
    print(f"Number of times {player} was not out: {count}")
```
